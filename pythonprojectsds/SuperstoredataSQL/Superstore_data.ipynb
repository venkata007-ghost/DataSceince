```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import mysql.connector as mysql

import os
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'Superstore_final_dataset.csv')
df
```

```python
df.drop_duplicates(subset = df.columns.difference(['Row_ID']), inplace = True)
```

```python
df.dropna(axis = 0, subset = ['Order_ID', 'Order_Date', 'Ship_Date', 'Ship_Mode', 'Customer_ID', 'Customer_Name', 'Segment', 'Country', 'City', 'State', 'Postal_Code', 'Region', 'Product_ID', 'Category', 'Sub_Category', 'Product_Name', 'Sales'], inplace = True)
```

```python
df
```

```python
df.dtypes
```

```python
# count the number of missing values in each row
missing_values_count = df.isnull().sum(axis = 1)

df = df[missing_values_count <= 2]
```

```python
df.dropna(subset = ['Order_Date', 'Ship_Date', 'Product_Name'], inplace = True)
df
```

```python
df = df[df['Category'].isin(['Furniture', 'Office Supplies','Technology'])]
df['Postal_Code'] = df['Postal_Code'].astype(str).str.extract('(\d{5})', expand = False)
df['Sales'] = df['Sales'].astype(float)
df['Order_Date'] = df['Order_Date'].astype('datetime64')
df['Ship_Date'] = df['Ship_Date'].astype('datetime64')
```

```python
df
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format = "%Y/%m/%d", errors = 'coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], format = "%Y/%m/%d", errors = 'coerce')
mean_order_date = np.datetime64(df['Order_Date'].mean())
df['Order_Date'] = df['Order_Date'].fillna(mean_order_date)
mean_ship_date = np.datetime64(df['Order_Date'].mean())
df['Ship_Date'] = df['Ship_Date'].fillna(mean_ship_date)
df
```

```python
df.dropna(axis = 0, subset = ['Order_Date', 'Ship_Date'], inplace = True)
df
```

```python
df = df[(pd.to_datetime(df['Ship_Date']) >= pd.to_datetime(df['Order_Date'])) & (df['Sales'] >= 0)]
df
```

```python
df['Postal_Code'].fillna(df['Postal_Code'].mode()[0], inplace = True)
df
```

```python
df.to_csv('superstore_final_data_cleaned.csv')
```

```python
import pymysql

# Connect to the database
host = "localhost"
database = "datasceince"
user = "root"
password = "password"

db = pymysql.connect(host = host, database = database, user = user, password = password)

cursor = db.cursor()
```

```python

# Create the "superstore" table
cursor.execute("DROP TABLE IF EXISTS superstore_dataset")
cursor.execute("""
    CREATE TABLE superstore_dataset (
        Row_ID INT NOT NULL,
        Order_ID VARCHAR(255),
        Order_Date DATE,
        Ship_Date DATE,
        Ship_Mode VARCHAR(255),
        Customer_ID VARCHAR(255),
        Customer_Name VARCHAR(255),
        Segment VARCHAR(255),
        Country VARCHAR(255),
        City VARCHAR(255),
        State VARCHAR(255),
        Postal_Code VARCHAR(255),
        Region VARCHAR(255),
        Product_ID VARCHAR(255),
        Category VARCHAR(255),
        Sub_Category VARCHAR(255),
        Product_Name VARCHAR(255),
        Sales FLOAT
    )
""")
```

```python
# Insert the data into the "superstore" table
for row in df1.itertuples():
    cursor.execute("""
        INSERT INTO superstore_dataset (
            Row_ID,
            Order_ID,
            Order_Date,
            Ship_Date,
            Ship_Mode,
            Customer_ID,
            Customer_Name,
            Segment,
            Country,
            City,
            State,
            Postal_Code,
            Region,
            Product_ID,
            Category,
            Sub_Category,
            Product_Name,
            Sales
        )
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, row[1:])

# Commit the changes and close the database connection
db.commit()
print("Data successfully inserted into the 'superstore_dataset' table!")
db.close()
```
